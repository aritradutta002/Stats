# Binomial Distribution

```python
# There are totally 50 ships in a port.If on an average 1 ship out of 10 do not arrive safely to ports. 

# i) Probability of getting exactly 3 ship to sink
# ii) Probability of getting atleast 3 ship to sink
# iii) Probability of getting atmost 3 ship to sink
# iv) Probability of getting between 3 and 10 ships to sink
# pmf
# cdf
```

```python
# i) Probability of getting exactly 3 ship to sink
n=50  
p=0.1
x=3
stats.binom.pmf(x,n,p)  # x,n,p
```

```python
# ii) Probability of getting atleast 3 ship to sink
1- stats.binom.cdf(2,n,p)
```

```python
# iii) Probability of getting atmost 3 ship to sink - [0-3]
stats.binom.cdf(3,n,p)
```

```python
# iv Probability of getting between 3 and 10 ships to sink -[4,5,6,7,8,9]
stats.binom.cdf(9,n,p) - stats.binom.cdf(3,n,p)
```

```python
# iv Probability of getting from 3 to 10 ships to sink -[3,4,5,6,7,8,9,10]
stats.binom.cdf(10,n,p) - stats.binom.cdf(2,n,p)
```

```python
pmf=[]
cdf=[]
for i in range(0,n+1):
    pmf.append(stats.binom.pmf(i,n,p))
    cdf.append(stats.binom.cdf(i,n,p))
```

```python
plt.plot(np.arange(0,n+1),pmf,'o-')
plt.show()
```

```python
plt.plot(np.arange(0,n+1),cdf,'o-')
plt.show()
```

# Z test

```python
# two tailed

# The mean life time of 50 electric bulbs produced by a manufacturing company
# is estimated to be 825 hours., If m is the mean life
# time of all the bulbs produced by the company .test the hypothesis that m = 900 hours , 
# with a standard deviation of 110 hours  at  5% level of significance.
```

```python
# Ho : m=900
# Ha : m!=900
z=(825-900)/(110/((50)**0.5))
z
```

```python
p_val=stats.norm.sf(np.abs(z))*2  # Z stat -> pval
print(p_val)
```

```python
(1-stats.norm.cdf(np.abs(z)))*2
```

```python
# Since the p_val is less than 0.05 we reject the null hypothesis.
```

```python
# one tailed:

# A sample of 400 bulbs is manufactued found to have a mean lifetime is 171.38 hrs. Can it
# reasonably be regarded as a large population with mean  more than 173 hrs
# and standard deviation of 3.3 hrs so that company can make profit(Test at 5% level), Will company make profit?
```

```python
# Ho: mu <=173 (no profit)
# Ha : mu > 173 (profit)
```

```python
x_bar = 171.38
n=400
mu=173
sigma =3.3
z= (x_bar-mu)/(sigma/(n**0.5))
p_val= stats.norm.sf(abs(z))
print(abs(z),p_val)

```

```python
# We reject the null hypothesis
```

# T test

```python
# two tail t test:

# A premium golf ball production line must produce all of its balls to 1.615 
#ounces in order to get the top rating (and therefore the top dollar). 
#Samples are drawn hourly and checked. If the production line gets out of 
#sync with a statistical significance of more than 1%, it must be shut down and
# repaired. This hour’s sample of 18 balls has a mean of 1.611 ounces 
#and a standard deviation of 0.065 ounces. Do you shut down the line?
```

```python
# Ho : w = 1.615
# ha : w != 1.615
```

```python
x_bar = 1.611
n=18
mu=1.615
s =0.065
df=n-1
t= (x_bar-mu)/(s/(n**0.5))
p_val= stats.t.sf(abs(t),df)*2
print(t,p_val)

```

```python
# one tail t test:

# An alternator manufacturer must produce its alternators
#so that they are 95% confident that it runs at less than 71.1o C 
#under stress test in order to meet the production 
#requirements for sale to the
# US government. The stress test is performed on random samples drawn 
#from the production line on a
# daily basis. Today’s sample of 7 alternators has a mean of 71.3o C 
# and a standard deviation of 0.214o C
# . Is there a production quality issue?
```

```python
# Ho : mu<71.1 
# Ha : mu>71.1
```

```python
x_bar = 71.3
n=7
mu=71.1
s =0.214
df=n-1
t= (x_bar-mu)/(s/(n**0.5))
p_val= stats.t.sf(abs(t),df)
print(t,p_val)
```

```python
# Unpaired t ttest:
Weight_loss_Male   = [ 3.69, 4.12, 4.65, 3.19,  4.34, 3.68, 4.12, 4.50, 3.70, 3.09,3.65, 4.73, 3.93, 3.46, 3.28, 4.43, 4.13, 3.62, 3.71, 2.92]
Weight_loss_Female = [2.99, 1.80, 3.79, 4.12, 1.76, 3.50, 3.61, 2.32, 3.67, 4.26, 4.57, 3.01, 3.82, 4.33, 3.40, 3.86]


# Do the the manual calculation and compare the results with inbuitl fucntion.
```

```python
male = np.array(Weight_loss_Male)
female = np.array(Weight_loss_Female)
nm = len(male)
nf = len(female)
xm = male.mean()
xf = female.mean()
sm = np.std(male,ddof = 1)
sf = np.std(female,ddof = 1)
se = np.sqrt((sm*sm/nm) + (sf*sf/nf))
tstat = ((xf - xm) - 0)/se
tcrit = stats.t.ppf(0.05/2,nm+nf-2)
pvalue = (stats.t.sf(tstat,nm+nf-2))*2
print(tstat,tcrit,pvalue)
```

```python
# Ho : WEight loss male = Wei loss of female
# Ho : WEight loss male != Wei loss of female
stats.ttest_ind(Weight_loss_Female,Weight_loss_Male)
```
