### For all Columns:

```python
passing_test=[]
for col in columns:
    p_val_nor=shapiro(data[col])[1]
    if p_val_nor>0.05:
        clas_1=data[data['Cultivator']==1][col]
        clas_2=data[data['Cultivator']==2][col]
        clas_3=data[data['Cultivator']==3][col]
        if f_oneway(clas_1,clas_2,clas_3)[1]<0.05:
            passing_test.append((col,True,'Anova'))
        else:
            passing_test.append((col,False,'Anova'))
    else:
        clas_1=data[data['Cultivator']==1][col]
        clas_2=data[data['Cultivator']==2][col]
        clas_3=data[data['Cultivator']==3][col]
        if kruskal(clas_1,clas_2,clas_3)[1]<0.05:
            passing_test.append((col,True,'Kruskal'))
        else:
            passing_test.append((col,False,'Kruskal'))
```
